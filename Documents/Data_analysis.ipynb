```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Perception_final.csv")
df = pd.DataFrame(df)
```

```python
# Example of multiple replacements
df = df.replace({
    'Very often': 'Very Often',
    'Occasionaly': 'Occasionally',
    '80 to 100 kmph': '70 to 100 kmph'
})
```

```python
import string
from itertools import product


columns = []
letters = list(string.ascii_uppercase)  # List of letters A to Z

for length in range(1, 3):  # To generate from 'B' to 'Z' and 'AA' to 'BU'
    columns.extend([''.join(comb) for comb in product(letters, repeat=length)])

columns = columns[1:]

df.columns = columns[:len(df.columns)]

```

```python
import textwrap

value_counts = df.loc[:, 'AO':'AT'].apply(pd.Series.value_counts)

# Plotting donut charts for each column
fig, axs = plt.subplots(3, 2, figsize=(15, 15))  # Adjust figure size as needed
fig.suptitle('Perception Regarding ADAS', fontsize=20)

# Flatten the axs array for easier iteration
axs = axs.flatten()

# Manual titles for each subplot
titles = [
    "Have you ever experienced driving a car with advanced driver assistance systems (ADAS)?",
    "Do you own a car with ADAS?",
    "How often do you drive the car equipped with ADAS?",
    "Do you think driving a car with ADAS would assist the driving operation?",
    "Do you feel driving a car with ADAS leads to engaging in non-driving activities, when the ADAS function is turned on?",
    "With the ADAS functions being able to produce both audio and video display warnings, which among the following would you consider more efficient to provide safety?"
]

# Manual labels for each column
manual_labels = {
    'AO': ["Yes", "No"],
    'AP': ["Yes", "No"],
    'AQ': ["Occasionally", "Rarely", "Never"],
    'AR': ["Yes", "No"],
    'AS': ["Yes", "No"],
    'AT': ["Audio + Video warning", "Audio warning (with beep sounds)", "Video display warning (with blinking light)"]
}

# Define a consistent color mapping
color_map = plt.cm.tab10.colors

# Plot each donut chart
for i, col in enumerate(value_counts.columns):
    # Calculate percentages, handling NaN values
    total = value_counts[col].sum()
    labels = manual_labels[col]
    
    # Ensure that percentages include all labels with zero counts if missing
    percentages = [value_counts[col].get(label, 0) / total * 100 for label in labels]

    # Get the corresponding colors
    pie_colors = color_map[:len(labels)]  # Select appropriate number of colors

    # Plotting the outer ring (larger pie chart)
    wedges1, texts1, autotexts1 = axs[i].pie(percentages, startangle=90, colors=pie_colors,
                                             wedgeprops={'width': 0.4, 'edgecolor': 'white'},
                                             autopct='%1.1f%%', pctdistance=0.85)

    # Plotting the inner ring (smaller pie chart)
    wedges2, _ = axs[i].pie([1], radius=0.6, colors=['white'], wedgeprops={'edgecolor': 'white'})

    # Wrap the title text and set it
    wrapped_title = textwrap.fill(titles[i], width=50)
    axs[i].set_title(wrapped_title)
    axs[i].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

    # Custom legend
    axs[i].legend(wedges1, labels, title='Responses', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
import pandas as pd

# Define a function to convert range strings to their average numeric value
def range_to_avg(range_str):
    if range_str == 'more than 120 kmph':
        return 120  # Assign a high value for 'more than 120 kmph'
    elif isinstance(range_str, str):
        range_str = range_str.replace(' kmph', '')
        parts = range_str.split(' to ')
        if len(parts) == 2:
            return (int(parts[0]) + int(parts[1])) / 2
    return None

# Apply the function to the relevant columns and create a new column 'Average_Ranges'
df['Average_Ranges'] = df[['Z', 'AA', 'AB', 'AC']].applymap(range_to_avg).mean(axis=1)

# Define a function to filter the responses in column 'AD'
def filter_accident_response(response):
    if response.startswith('Yes'):
        return 'Yes'
    elif response.startswith('No'):
        return 'No'
    return None

# Apply the function to create the new column 'Accident_Previously'
df['Accident_Previously'] = df['AD'].apply(filter_accident_response)

# Now, let's compare the average driving speed for those who have been in an accident vs those who haven't
avg_speed_by_accident = df.groupby('Accident_Previously')['Average_Ranges'].mean()

# Plotting
plt.figure(figsize=(8, 5))  # Adjust figure size as needed
avg_speed_by_accident.plot(kind='bar')
plt.title('Average Driving Speed by Accident History')
plt.xlabel('Accident Previously')
plt.ylabel('Average Driving Speed (kmph)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create bins for commuting time
def categorize_commute_time(time_str):
    if 'Less than 30 min' in time_str:
        return '<30 min'
    elif '30 min to 1 hour' in time_str:
        return '30 min to 1 hour'
    elif '1 hour to 2 hours' in time_str:
        return '1-2 hours'
    elif '2 hours to 3 hours' in time_str:
        return '2-3 hours'
    elif '3 hours to 4 hours' in time_str:
        return '3-4 hours'
    elif 'More than 4 hours' in time_str:
        return '>4 hours'
    return 'Unknown'

df['Commute_Time'] = df['BT'].apply(categorize_commute_time)

# Explicitly define the order of the categories
categories = ['<30 min', '30 min to 1 hour', '1-2 hours', '2-3 hours', '3-4 hours', '>4 hours']
df['Commute_Time'] = pd.Categorical(df['Commute_Time'], categories=categories, ordered=True)

# Define a function to count the types of autonomous vehicles preferred
def count_autonomous_preferences(pref):
    if 'I do not want to travel in an autonomous vehicle' in pref:
        return 0
    return 1

# Apply the function to create a new column with the count of preferences
df['Autonomous_Preference'] = df['BH'].apply(count_autonomous_preferences)

# Group by commute time and calculate the average preference
commute_vs_preference = df.groupby('Commute_Time')['Autonomous_Preference'].mean().sort_index()*100

# Plot the comparison
plt.figure(figsize=(10, 6))
commute_vs_preference.plot(kind='bar', color='skyblue', legend=False)
plt.title('Impact of Commute Time on Desire for Autonomous Vehicles')
plt.xlabel('Commute Time')
plt.ylabel('Percentage Preferring Autonomous Vehicles')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
